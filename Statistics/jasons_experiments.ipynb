```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Given coordinates
x_data = np.array([10, 50, 100, 150])
y_data = np.array([628, 294, 268, 266])

# Plot the given coordinates
plt.scatter(x_data, y_data, color='red', label='Data points')

# Define the exponential decay function for regression
def exp_decay_func(x, a, b):
    return a * np.exp(-b * x)

# Define the exponential function for regression
# def exp_func(x, a, b):
#     return a * np.exp(b * x)

# Perform exponential regression
params, covariance = curve_fit(exp_decay_func, x_data, y_data)

# Extract the parameters
a, b = params

# Generate x values for plotting the regression line
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = exp_decay_func(x_fit, a, b)


# Plot the regression line
plt.plot(x_fit, y_fit, label=f'Exponential decay fit: y = {a:.2f} * exp(-{b:.2f} * x)')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Regression')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Print the regression equation
print(f'The exponential decay regression equation is: y = {a:.2f} * exp(-{b:.2f} * x)')
```

There are 3 players.
Each player must roll at least once, but no more than twice.
Each player will not choose to roll a second time if their first roll is a 20.
Each player will choose to roll a second time if their first roll is less than 20 and they predict that their total rolls will be larger than or equal to the other player's total rolls.
If a player's total rolls exceed 20, that player will not advance.
Only those players who have the highest total rolls will advance.  In the event that there is a tie, all those involved in the tie will advance.

The players will roll in the fillowing sequence:

1. The first player rolls his first roll.
2. The first player decided whether to roll a second time.
3. If the first player chooses to roll a second time, he will do so.

4. The second player rolls his first roll.
5. The second player decides whether to roll a second time.
6. If the second player chooses to roll a second time, he will do so.

7. The third player rolls his first roll.
8. The third player decides whether to roll a second time.
9. If the third player chooses to roll a second time, he will do so.
