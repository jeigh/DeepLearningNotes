```python
import numpy as np
import matplotlib.pyplot as plt
```

### Normal Distribution 
A continuous probability distribution that is symmetric about the mean.  Also called the Gaussian distribution.

$$ 
P(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} 
$$

where:

* $ P(x) $ is the probability density function,
* $ \mu $ is the mean of the distribution,
* $ \sigma $ is the standard deviation of the distribution,
* $ e $ is the base of the natural logarithm (approximately equal to 2.71828).


```python
# Define the Gaussian (Normal) distribution function
def gaussian(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Parameters for the Gaussian distribution
mu = 0      # Mean
sigma = 1   # Standard deviation

# Generate a range of x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Compute the Gaussian distribution values for these x values
y = gaussian(x, mu, sigma)

# Plot the Gaussian distribution
plt.plot(x, y)
plt.title('Gaussian (Normal) Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

### Power Law Distribution
A continuous probability distribution that describes the probability distribution of a random variable.  Also called the Pareto distribution.

$$  P(x) = C x^{-\alpha} $$
where:

* $ P(x) $ is the probability of the random variable $ x $,
* $ C $ is a normalization constant,
* $ \alpha $ is a positive parameter that characterizes the distribution.

```python
# Define the Power Law distribution function
def power_law(x, alpha, C):
    return C * x ** (-alpha)

# Parameters for the Power Law distribution
alpha = 2.5  # Positive parameter that characterizes the distribution
C = 1        # Normalization constant

# Generate a range of x values
x = np.linspace(1, 10, 1000)  # x should be greater than 0

# Compute the Power Law distribution values for these x values
y = power_law(x, alpha, C)

# Plot the Power Law distribution
plt.plot(x, y)
plt.title('Power Law Distribution')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.yscale('log')  # Use a logarithmic scale for better visualization
plt.grid(True)
plt.show()
```

### Uniform Distribution: 
A continuous probability distribution that is defined by two parameters: the minimum and maximum values. Also called the rectangular distribution.

$$ 
P(x) = \frac{1}{b-a}
$$

where:

* $ P(x) $ is the probability density function,
* $ a $ is the minimum value,
* $ b $ is the maximum value.


```python
# Define the Uniform distribution function
def uniform(x, a, b):
    return np.where((x >= a) & (x <= b), 1 / (b - a), 0)

# Parameters for the Uniform distribution
a = 0  # Minimum value
b = 1  # Maximum value

# Generate a range of x values
x = np.linspace(a - 1, b + 1, 1000)

# Compute the Uniform distribution values for these x values
y = uniform(x, a, b)

# Plot the Uniform distribution
plt.plot(x, y)
plt.title('Uniform Distribution')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.ylim(0, 1.2)  # Set y-axis limits for better visualization
plt.grid(True)
plt.show()
```

### Chi-Squared Distribution
A continuous probability distribution that describes the sum of squared standard normal deviates.

$$
P(x) = \frac{1}{2^{k/2} \Gamma(k/2)} x^{(k/2)-1} e^{-x/2}
$$

where:

* $ P(x) $ is the probability density function,
* $ k $ is the degrees of freedom,
* $ \Gamma $ is the Gamma function,
* $ e $ is the base of the natural logarithm (approximately equal to 2.71828).

```python
from scipy.stats import chi2

# Parameters for the Chi-Squared distribution
k = 4  # Degrees of freedom

# Generate a range of x values
x = np.linspace(0, 20, 1000)

# Compute the Chi-Squared distribution values for these x values
y = chi2.pdf(x, k)

# Plot the Chi-Squared distribution
plt.plot(x, y)
plt.title('Chi-Squared Distribution')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.grid(True)
plt.show()
```

### Student's T-distribution
A continuous probability distribution that describes the mean of a normally distributed population when the sample size is small.  

$$ 
P(t) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu \pi} \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu + 1}{2}} 
$$

where:

* $ P(t) $ is the probability density function,
* $ \nu $ is the degrees of freedom,
* $ \Gamma $ is the Gamma function,
* $ t $ is the variable.

```python
from scipy.stats import t

# Parameters for the Student's t-distribution
df = 5  # Degrees of freedom

# Generate a range of x values
x = np.linspace(-10, 10, 1000)

# Compute the Student's t-distribution values for these x values
y = t.pdf(x, df)

# Plot the Student's t-distribution
plt.plot(x, y)
plt.title("Student's t-Distribution")
plt.xlabel('x')
plt.ylabel('P(x)')
plt.grid(True)
plt.show()
```

### F-Distribution
A continuous probability distribution that describes the ratio of two independent chi-squared distributions.

$$
P(F) = \frac{\left(\frac{d_1}{d_2}\right)^{d_1/2} \left(F\right)^{(d_1/2)-1}}{B\left(\frac{d_1}{2}, \frac{d_2}{2}\right) \left(1 + \frac{d_1}{d_2}F\right)^{(d_1+d_2)/2}} 
$$ 

where:

* $ P(F) $ is the probability density function,
* $ d_1 $ and $ d_2 $ are the degrees of freedom for the numerator and the denominator, respectively,
* $ B $ is the Beta function,
* $ F $ is the variable.

```python
from scipy.stats import f

# Parameters for the F-distribution
d1 = 5  # Degrees of freedom for the numerator
d2 = 2  # Degrees of freedom for the denominator

# Generate a range of x values
x = np.linspace(0, 5, 1000)

# Compute the F-distribution values for these x values
y = f.pdf(x, d1, d2)

# Plot the F-distribution
plt.plot(x, y)
plt.title('F-Distribution')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.grid(True)
plt.show()
```

### Log-Normal Distribution 
A continuous probability distribution that is the logarithm of a normal distribution.  Also called the Galton distribution.
$$
P(x) = \frac{1}{x \sigma \sqrt{2\pi}} e^{-\frac{(\ln(x)-\mu)^2}{2\sigma^2}}
$$
where:

* $ P(x) $ is the probability density function,
* $ \mu $ is the mean of the natural logarithm of the variable,
* $ \sigma $ is the standard deviation of the natural logarithm of the variable,
* $ \ln(x) $ is the natural logarithm of ( x ),
* $ e $ is the base of the natural logarithm (approximately equal to 2.71828).


```python
from scipy.stats import lognorm

# Parameters for the Log-Normal distribution
mu = 0       # Mean of the underlying normal distribution
sigma = 0.954  # Standard deviation of the underlying normal distribution

# Generate a range of x values
x = np.linspace(0, 5, 1000)

# Compute the Log-Normal distribution values for these x values
y = lognorm.pdf(x, s=sigma, scale=np.exp(mu))

# Plot the Log-Normal distribution
plt.plot(x, y)
plt.title('Log-Normal Distribution')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.grid(True)
plt.show()
```
